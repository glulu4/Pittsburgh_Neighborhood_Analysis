```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

db = pd.read_csv("Natural_Environment_Conditions_2010.csv", sep=",")

avg_num_trees = db["# Street Trees"].mean()
avg_num_parkSpace = db["Park Space (acres/1000 pers.)"].mean()


mask = (db["# Street Trees"] > avg_num_trees  ) & (db["Park Space (acres/1000 pers.)"] > avg_num_parkSpace - 20 )
maskedDB = db[mask] 

top30_env = maskedDB[["Neighborhood", "# Street Trees", "Park Space (acres/1000 pers.)" ]].sort_values("# Street Trees", ascending = False)
top30_env
```

# Conclusion

#### The above code reads in a csv file then calculates the average of two columns ( street trees and park space per 1000 people ). After that, a mask is applied to the data to filter out neighborhoods that don't meet the threshholds of the averages. Then, those two coloumns are selected and are sorted by the number of street trees.
